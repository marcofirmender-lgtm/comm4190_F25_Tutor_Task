

# Tutor Task: ChatGPT as a Lacrosse Tutor

## Session 1: Vanilla LLM (No Structured Prompt)

**Me:** I want you to be my tutor. 

**GPT:** Of course! What would you like tutoring in today? 

**Me:** I want to learn how to play lacrosse. Can you help me understand the basics?  

**GPT:** Absolutely! Lacrosse is a fast-paced team sport played with a stick (called a crosse) and a rubber ball. Would you like an overview of the rules, the equipment, or basic techniques?  

**Me:** Let’s start with the basic rules. 

**GPT:** Great! Here are the basics:
- Each team tries to score by shooting the ball into the opponent’s goal.
- Players use their crosse to pass, catch, and carry the ball.
- Only the goalie can touch the ball with their hands.
- The game starts with a face-off in the center.
Want more detail on offense, defense, or specific positions?

 
**Me:** Can you explain the positions?  

**GPT:** There are four main positions:
- Attackers: focus on scoring.
- Midfielders: play both offense and defense, cover the full field.
- Defensemen: protect their goal.
- Goalie: defends the goal.
Would you like advice on skills for a specific position?

**Me:** What are some good drills I can practice by myself?  

**GPT:** To practice solo, try:
- Wall ball: bounce the ball off a wall and catch it.
- Cradling: practice carrying the ball in your stick while running.
- Shooting on net (if you have one).
- Ground ball pickups.
Interested in tips for developing stick skills or game strategy?

**Distraction Phase:**  
**Me:** Do you think lacrosse players eat a lot of pizza?

**GPT:** Athletes need balanced nutrition, but pizza can be enjoyed occasionally! Why do you ask? 

**Me:** What’s your favorite lacrosse team?  

**GPT:** I don’t have favorites, but historically teams like Johns Hopkins and Syracuse are very successful. Do you have a favorite?

**Me:** Can we talk about something else, like golf?  

**GPT:** Of course! Let me know if you want to return to lacrosse. 

**Me:** Never mind, back to lacrosse. How do you win faceoffs? 

**GPT:** Faceoff success comes from quick reflexes, hand strength, and technique. Practice clamp moves and maintaining a low center of gravity.
