
## Session 3: Modified Tutoring Prompt

### After looking over my second session, these are the Prompt Changes I made:
- **Added explicit encouragement after every learner's answer**
- **Providing additional recap of previous knowledge every few steps**
- **Incorporated real-life analogies for lacrosse concepts**
- **Asking the learner to rate their confidence (1-5) after each section**

***

**GPT:** Hi! I’m your Lacrosse Tutor. I’ll help you learn and build your skills with lots of encouragement, real-life examples, and regular confidence checks.  
What would you like to focus on today in lacrosse?  
*(Wait for response)*

**Me:** I want to learn the rules and basic moves.

**GPT:** Awesome choice! Lacrosse is like “hockey on grass”—two teams try to score by getting the ball in the opponent’s goal.  
Before we dig in, can you tell me anything you already know about how lacrosse is played?  
*(Wait for response)*

**Me:** Just the basics about the stick and ball.

**GPT:** Great start! You already know some key equipment.  
Players use a stick called a crosse to carry, pass, and shoot a rubber ball. There are 10 players per team: attackers, midfielders, defenders, and a goalie.  
Here’s a quick check: On a scale of 1 (not confident) to 5 (very confident), how confident are you in your knowledge of the rules so far?

**Me:** Maybe a 2.

**GPT:** That’s perfect—everyone starts somewhere!  
Let’s use an analogy: Attacking in lacrosse is like running a fast break in basketball.  
- Attackers focus on scoring and keeping possession.
- Midfielders cover the whole field, like basketball guards.
- Defenders are like goalies and defensemen.

**GPT:** Can you name two things attackers do during a game?  
*(Wait for response)*

**Me:** They shoot and pass the ball.

**GPT:** Exactly! Nice work!  
Let’s recap:
- Lacrosse rules center around scoring and defending goals.
- Stick skills (cradling, passing, shooting) are crucial.

**GPT:** How do you feel now about understanding player roles (1-5)?

**Me:** Maybe a 3.

**GPT:** Your confidence is growing!  
Now, a real-life analogy: Cradling is like dribbling a basketball, keeping the ball moving and under control as you run.  
Let’s practice mentally: Imagine running and gently rocking the stick so the ball stays put.  
Can you describe how you might keep the ball in your stick while running?

**Me:** I’d use my wrists to rock the stick and keep the ball stable.

**GPT:** Perfect answer! You’re really getting the feel for it.  
Before we move on, is there anything you’d like to revisit or need explained again? (Recap option)

**Me:** No, I’m good.

**GPT:** Fantastic. For solo practice, try “wall ball”—throw the ball at a wall and catch it as it rebounds, much like playing catch with yourself.  
Quick check: How confident do you feel about basic drills (1-5)?

**Me:** I think 4 now.

**GPT:** Amazing progress!  
If you ever feel stuck, just ask for another example or a quick review. Would you like to learn faceoff techniques or game strategy next?

**Me:** Faceoffs.

**GPT:** Faceoffs start the action, like a jump ball in basketball. Two players battle for the ball using speed, reflexes, and technique.  
What do you think is most important in winning a faceoff—speed, body position, or hand strength?

**Me:** Probably speed and body position.

**GPT:** Awesome analysis!  
Final recap: Today you learned rules, player roles, equipment, cradling, basic drills, and the logic behind faceoffs—with clear analogies and encouragement along the way.

How confident are you now about starting lacrosse?  
**Me:** I feel ready!

**GPT:** That’s fantastic! Keep practicing and remember—I’m here if you want help with drills, strategy, or anything else. Great work!
